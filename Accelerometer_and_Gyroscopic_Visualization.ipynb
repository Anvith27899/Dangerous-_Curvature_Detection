#Accelerometer and Gyroscopic Visualization

```python
from google.colab import files
uploaded=files.upload()
```

```python
import pandas as pd
import io
df=pd.read_excel(io.BytesIO(uploaded['Raw Data.xlsx']))
print(df.head())
```

```python
d=pd.DataFrame(df)
x=d.iloc[:,1].values
y=d.iloc[:,2].values
z=d.iloc[:,3].values
print(x,"\n",y,"\n",z)
```

```python
import numpy as np
import matplotlib.pyplot as pl
```

If we consider only starting 20 rows

```python
x1=d.iloc[:20,1].values
y1=d.iloc[:20,2].values
z1=d.iloc[:20,3].values
```

```python
fig=pl.figure()
a=pl.axes(projection='3d')
a.plot3D(x1,y1,z1,c='black')
a.scatter3D(x1,y1,z1,c=z1,cmap='Greens')
pl.title("Accelerometer Data")
```

If we consider all values

```python
a=pl.axes(projection='3d')
a.plot3D(x,y,z,c='black')
a.scatter3D(x,y,z,c=z,cmap='Greens')
```
