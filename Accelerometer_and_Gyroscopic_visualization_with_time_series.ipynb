For Accelerometer Data

```python
from google.colab import files
uploaded=files.upload()
```

```python
import pandas as pd
import io
df=pd.read_excel(io.BytesIO(uploaded['Raw Data.xlsx']))
print(df.head())
```

```python
d=pd.DataFrame(df)
t=d.iloc[:,0].values
x=d.iloc[:,1].values
y=d.iloc[:,2].values
z=d.iloc[:,3].values
print(t,"\n",x,"\n",y,"\n",z)
```

```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
pl.plot(t,x,c='red')
pl.legend("X wrt Time")
pl.xlabel("Time")
pl.ylabel("X")
```

```python
pl.plot(t,y,c='blue')
pl.legend("Y wrt Time")
pl.xlabel("Time")
pl.ylabel("Y")
```

```python
pl.plot(t,z,c='green')
pl.legend("Z wrt Time")
pl.xlabel("Time")
pl.ylabel("Z")
```
